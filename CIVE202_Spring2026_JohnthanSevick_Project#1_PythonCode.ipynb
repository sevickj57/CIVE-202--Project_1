# TASK 1

```python
#import pandas and make data dataframe
import pandas as pd
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')
air_data = pd.DataFrame(air_data)
air_data.head()
```

```python
#TASK 1
#What are the 5 locations in Nebraska with the highest mean and median concentrations of VOC, PM 2.5 and PM 10.0
#voc top5 mean
voc_data1=air_data.groupby('sensor.name').agg(Mean_voc =('voc','mean'))
task1_voc_mean=voc_data1.sort_values(by = 'Mean_voc', ascending = False).head(5)
task1_voc_mean
```

```python
#voc top5 median
voc_data2=air_data.groupby('sensor.name').agg(Median_voc =('voc','median'))
task1_voc_median=voc_data2.sort_values(by = 'Median_voc', ascending = False).head(5)
task1_voc_median
```

```python
#pm2.5 top 5 mean
pm25_data1=air_data.groupby('sensor.name').agg(Mean_pm25 =('pm2.5_atm','mean'))
task1_pm25_mean=pm25_data1.sort_values(by = 'Mean_pm25', ascending = False).head(5)
task1_pm25_mean
```

```python
#pm2.5 top 5 median
pm25_data2=air_data.groupby('sensor.name').agg(Median_pm25 =('pm2.5_atm','median'))
task1_pm25_median=pm25_data2.sort_values(by = 'Median_pm25', ascending = False).head(5)
task1_pm25_median
```

```python
#pm10.0 top 5 mean
pm10_data1=air_data.groupby('sensor.name').agg(Mean_pm10 =('pm10.0_atm','mean'))
task1_pm10_mean=pm10_data1.sort_values(by = 'Mean_pm10', ascending = False).head(5)
task1_pm10_mean
```

```python
#pm10.0 top 5 median
pm10_data2=air_data.groupby('sensor.name').agg(Median_pm10 =('pm10.0_atm','median'))
task1_pm10_median=pm10_data2.sort_values(by = 'Median_pm10', ascending = False).head(5)
task1_pm10_median
```

```python
#Does humidity and temperature have a noticeable effect on air quality
#set catagories for temperature
def temp_cat(t):
    if t < 32:
        return 'Below Freezeing'
    elif 32 <= t <= 50:
        return 'Cool'
    elif 51 <= t <= 70:
        return 'Warm'
    else:
        return 'Hot'
#set catagories for humidity
def humid_cat(h):
    if h < 50:
        return "Low Humidity"
    elif h < 80:
        return "High Humidity"
    else:
        return "Very High Humidity"
#apply catagorizations 
air_data['temperature.level'] = air_data['temperature'].apply(temp_cat)
air_data['humidity.level'] = air_data['humidity'].apply(humid_cat)
```

```python
#group temp catagories 
temp_data1=air_data.groupby('temperature.level').agg(voc_mean = ('voc','mean'),
                                                     pm25_mean = ('pm2.5_atm','mean'),
                                                     pm10_mean = ('pm10.0_atm','mean'))
temp_data1
```

```python
humid_data1=air_data.groupby('humidity.level').agg(voc_mean = ('voc','mean'),
                                                     pm25_mean = ('pm2.5_atm','mean'),
                                                     pm10_mean = ('pm10.0_atm','mean'))
humid_data1
```
